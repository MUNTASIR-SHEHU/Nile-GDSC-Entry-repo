```python
#Imports all required modules
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Task 1: Image Basics

```python
#Reads the image in the folder and returns as an array
image = cv2.imread("img.png")
```

```python
#Converts the image from BGR to RGB to be displayed properly
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#Takes the image again and coverts it from BGR to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Returns a shape tuple with 3 values and lables each value h, w, & c respectively
h, w, c = image_rgb.shape
#created a variable as a formatted stringed to be used later
dim_text = f"Dimensions: {h} x {w} x {c}"
```

```python
#Creates an area of width 10 and height 4 for the images to be displayed
plt.figure(figsize=(10,4))
#Divides are into 1 row and 2 columns and select left column
plt.subplot(1,2,1)
#Shows original image in rgb on selected column and titles it "Original Image"
plt.imshow(image_rgb)
plt.title("Original Image")
#hides axis ticks and labels
plt.axis("off")
#displays the dimensions in the selected x & y locations, sets a background colour, font size & text colour of black, 10 & white respectively
plt.text(10, -25, dim_text, color='white', fontsize=10, backgroundcolor='black')

#Divides are into 1 row and 2 columns and select right column
plt.subplot(1,2,2)
#Shows grayscale image with a colour map of gray on selected column and titles it "Grayscale Image"
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
#hides axis ticks and labels
plt.axis("off")

#adjusts spacing between subplots so titles and labels don’t overlap and renders the image
plt.tight_layout()
plt.show()
```
